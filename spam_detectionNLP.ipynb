```python
df=pd.read_csv('spam_ham.csv')
df
```

```python
def clean(text):
    text=text.lower()
    text=re.sub(r'[^\w\s]','',text)
    tokens=word_tokenize(text)
    tokens=[WordNetLemmatizer().lemmatize(word) for word in tokens]
    tokens=[word for word in tokens if word not in stopwords.words('english')]
    return "".join(tokens)
```

```python
df['email']=df['email'].apply(clean)
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(df['email'],df['label'],test_size=0.2,random_state=42,stratify=df['label'])
```

```python
pipeline=Pipeline([('tfidf',TfidfVectorizer()),('Classifier',MultinomialNB())])
```

```python
pipeline.fit(X_train,y_train)
```

```python
user_email=input("Enter a email:")
cleaned_email=clean(user_email)
predict=pipeline.predict([cleaned_email])
print("Result",predict[0])
```
